# 1. <a id='toc1_'></a>Aufgabenstellung   [&#8593;](#toc0_)
Dieses Projekt soll untersuchen, welche Erkenntnisse die Drogeriekette Rossmann aus ihren historischen Verkaufs- und Werbedaten, Daten zu Schulferien und Feiertagen sowie Wettbewerberdaten gewinnen kann. Das Projekt soll die Frage adressieren, wie sich diese Daten nutzen lassen, um den Betrieb und das Management der Filialen zu optimieren, mit dem Ziel, mehr Umsatz zu generieren. Darüber hinaus soll ermittelt werden, wie sich diese Datensätze einsetzen lassen, um die wöchentlichen Verkäufe (Umsatzerlöse) für jede Filiale mit einem ausreichenden Genauigkeitsniveau vorherzusagen.

## 1.1. <a id='toc1_1_'></a>Hintergrund   [&#8593;](#toc0_)

Rossmann betreibt über 4.300 Drogeriemärkte in 9 europäischen Ländern (vgl. https://de.wikipedia.org/wiki/Rossmann_(Handelskette)). Derzeit sind Rossmann-Filialleiter damit beauftragt, ihre wöchentlichen Verkaufsprognosen bis zu acht Wochen im Voraus zu erstellen. Die Verkaufsergebnisse der Filialen werden von vielen Faktoren beeinflusst, einschließlich Werbeaktionen, Wettbewerbsintensität, Schulferien und staatlichen Feiertagen, saisonalen Veränderungen und Standortbedingungen. Daher ist das Data Science Team des Unternehmens auf einer neuen Mission, eine vereinheitlichte Modellierungsmethode für Filialleiter zu erstellen, um die wöchentlichen Ergebnisse mit höherer Genauigkeit vorherzusagen. Das Management Team benötigt zudem auch einen Gesamtbericht mit machbaren bzw. umsetzbaren Strategien, um die allgemeine Leistung (Performance) aller Filialen zu verstehen und eine Möglichkeit zur Optimierung zukünftiger Verkaufsleistungen (d. h. der Umsatzerlöse) zu finden. Zuletzt muss jedem Filialleiter ein individueller Bericht zur Leistung der jeweiligen Filiale zur Verfügung gestellt werden.

## 2.1. <a id='toc2_1_'></a>Daten einlesen   [&#8593;](#toc0_)

Die automagischen Datentypen, die pandas beim Einlesen mit den Standardparametern für die vorgegebenen Dateien auswählt, sind aus verschiedenen Gründen ungeeignet für eine weitere Verarbeitung, unabhängig davon, wofür wir die Daten später benutzen werden (Businessanalyse oder Umsatzvorhersage).

Deshalb passen wir die Datentypen direkt beim Einlesen an und setzen für die einzelnen Dataframes entsprechend geeignete Indizes, um Speicherplatz zu sparen und allgemein für die Weitervearbeitung geeignete Dataframes zu erhalten.

### 2.1.1. <a id='toc2_1_1_'></a>train.csv   [&#8593;](#toc0_)
Beim Einlesen von ``train.csv`` parsen wir Ganzzahlen jeweils in den kleinsten Integertyp (``int16`` oder sogar ``int8``), der die Werte der Spalte aufnehmen kann, um Speicherplatz zu sparen, da es über eine Millionen Einträge gibt und der Dataframe verhältnismäßig groß werden und entsprechend viel Speicherplatz einnehmen kann. Die Spalte ``StateHoliday`` geben wir als ``category`` an. ``Date`` parsen wir direkt als datetime-Objekte und setzen ``Store`` und ``Date`` als Multiindex, da jedes (``Store``, ``Date``)-Paar eine Beobachtung darstellt. Durch diese Maßnahmen können wir den Speicherbedarf des Dataframes von anfangs 175 MB auf 15 MB senken. Außerdem fügen wir ein neues boolsches Feature ``isHoliday`` hinzu, das immer dann True ist, wenn für die Filiale und das Datum ein Feiertag ist, ansonsten False.

### 2.1.2. <a id='toc2_1_2_'></a>store.csv   [&#8593;](#toc0_)
Beim Einlesen von ``store.csv`` parsen wir die meisten Spalten als ``int32`` und ``StoreType``, ``Assortment`` und ``PromoIntervall`` als ``category``. ``CompetitionOpenSinceMonth`` und ``CompetitionOpenSinceYear`` fassen wir zu ``CompetitionSince`` zusammen, wobei wir der Einfachheit halber und aus Ermangelung genauerer Daten den Anfang des Monats von ``CompetitionOpenSinceMonth`` als Öffnungsdatum nehmen; ebenso verfahren wir mit ``Promo2SinceWeek`` und ``Promo2SinceYear``, die wir zu ``Promo2Since`` zusammenfassen und den Begin (Montag) der Woche von ``Promo2SinceWeek`` als Startdatum nehmen und wandeln die entstehenden Spalten in datetime-Objekte um. Die Spalten ``CompetitionOpenSinceMonth``, ``CompetitionOpenSinceYear``, ``Promo2SinceWeek`` und ``Promo2SinceYear`` löschen wir.

### 3.4.3. <a id='toc3_4_3_'></a>Sales und Customers   [&#8593;](#toc0_)

Stichprobenartig lässt sich feststellen, dass diese Tage ohne Umsätze anscheinend oft auftreten, wenn eine Filiale zum ersten Mal wieder öffnet, nachdem sie für ein oder zwei Wochen geschlossen war. 

Möglicherweise handelt es sich um Soft-Openings, nachdem die Filiale umgebaut wurde oder andere Änderungen am Filialbetrieb vorgenommen wurden. Ansonsten könnte es darauf hindeuten, dass Wiedereröffnungen von den Kunden nicht ausreichend wahrgenommen werden.

# 4. <a id='toc4_'></a>Modellbildung zur Umsatzvorhersage    [&#8593;](#toc0_)

Zunächst müssen wir uns überlegen, ob und wie wir die verfügbaren Daten überhaupt für eine sinnvolle Vorhersage nutzen können. Einige Features der Daten sind für eine wochenbasierte filialspezifische Vorhersage aus verschiedenen Gründen unbrauchbar. Andere Features sind essentiell für eine fundierte Vorhersage, während wieder andere Merkmale nur unter bestimmten Bedingungen interessant für die Vorhersage sind.

## 4.1. <a id='toc4_1_'></a>Merkmale die keinen Einfluss auf die Vorhersage haben    [&#8593;](#toc0_)

* DayOfWeek
  * Da wir sowieso auf Wochenbasis aggregieren und nur volle Wochen betrachten, hat DayOfWeek keinen Einfluss auf die Vorhersage, da jede (volle) Woche die gleiche Summe der Wochentage hat.
* StoreType, Assortment, CompetitionDistance und PromoInterval
  * Für einen einzelnen Store ändern sich diese Features nicht und es ist eigentlich auch nicht interessant, wie andere Filialen mit den gleichen Features bisher gelaufen sind. Selbst unter Einbeziehung aller verfügbaren hier nicht berücksichtigten Parameter, können den Daten nach ähnliche Filialen völlig verschiedene Umsatzverläufe haben. Letztendlich hängen die eigenen Umsätze einer Filiale ausschließlich von den eigenen historischen Daten ab.

## 4.2. <a id='toc4_2_'></a>Merkmale die Einfluss auf die Vorhersage haben (können)    [&#8593;](#toc0_)
* Open
  * Da wir auf Wochenebene arbeiten, ist Open ein guter Indikator, wie oft der Laden geöffnet hat(te), bzw. wie oft er in einer Woche geschlossen war, z.B. wegen Feiertagen, verkaufsoffenen Sonntagen o.ä. Je mehr Öffnungstage pro Woche (je weniger Schließungstage) desto höher der Umsatz.
* Promo
  * Gibt den Rhythmus an, in dem unternehmensweit die Werbeaktion Promo stattgefunden hat. In Promowochen bzw. an Promotagen wird allgemein mehr Umsatz gemacht, als an Nicht-Promotagen.
* StateHoliday (bzw. isHoliday) und SchoolHoliday
* CompetitionSince ist nur interessant, wenn der Zeitpunkt im Trainingszeitraum liegt
    * egal wie sich die Konkurenz auf den individuellen Umsatz auswirkt, gibt der Zeitpunkt an dem die Konkurenz geöffnet hat, einen Anhaltspunkt für Änderungen im Umsatzverhalten (Zunahme, Abnahme, Stagnation)
* PromoSince ist wie CompetitionSince nur dann interessant, wenn der Zeitpunkt im Trainingszeitraum liegt

## 4.4. <a id='toc4_4_'></a>Modelle [&#8593;](#toc0_)

Da die Daten von Filiale zu Filiale extrem variieren können, kann man kein Universalmodell entwickeln, das für alle Filialen gleich gut funktioniert. Auch eine ausführliche Kreuzvalidierung und langwieriges Ausloten der einzelnen Modelle und ihrer Hyperparameter kommt nicht infrage, da die Erstellung der Vorhersagen in Echtzeit erfolgen soll. 

Deshalb habe ich mich entschlossen, verschiedene ML-Algorithmen mit ihren jeweiligen Standardparametern zu testen und dann das beste Modell auszuwählen. In der Praxis funktioniert das innerhalb von Sekunden, im Gegensatz zu Minuten oder gar Stunden, die eine ausgewachsene Gittersuche benötigt.

Im einzelnen kommen folgende Modelle zum Einsatz bzw. werden in jedem Durchlauf getestet:
 * LinearRegression
 * Lasso
 * RidgeRegression
 * RandomForest
 * GradientBoosting
 * MLP
 * ARIMA, SARIMA
 * FB Prophet

 Der Code für das Perceptron ist standardmäßig auskommentiert, da das Fitting zu lange dauert, kann aber bei Bedarf aktiviert werden, falls die anderen Modelle keine befriedigenden Ergebnisse liefern.

 ARIMA und SARIMA sind Modelle zur Zeitreihenanalyse und tatsächlich Überbleibsel von einigen Experimenten, die ich nicht weiter verfolgt habe. Aber auch mit den gegenwärtigen Einstellungen liefern die beiden Modelle für manche Filialen erstaunlich gute Ergebnisse. Deswegen sind sie weiter im Code vorhanden.

 FaceBook Prophet ist ein weiteres Modell zur Zeitreihenanalyse, das viele der fummeligen Einstellungen und Berechnungen von Trends, Saisonalitäten, Residuen usw. automagisch vornimmt und oft mit wenigen Hilfestellungen relativ gute Ergebnisse produziert.

## 4.5. <a id='toc4_5_'></a>Bewertung der Modelle [&#8593;](#toc0_)

Um die Modelle bzw. ihre Vorhersagen bewerten und vergleichen zu können, setzte ich vier verschiedene Metriken ein:

* Mean Absolute Error (mae)
* Mean Absolute Percentage Error (mape)
* Root Mean Squared Error (rmse)
* R2 (r2)

Alle Metriken haben Vor- und Nachteile und lassen sich unterschiedlich intuitiv interpretieren. Für die endgültige Wahl des besten Modells nutze ich die Wurzel des mittleren quadratischen Fehlers (RMSE).
